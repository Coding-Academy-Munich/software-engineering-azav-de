### Zu Schritt 2: `JsonStore`

Die grundlegende Strategie: bei jeder Operation die gesamte Datei laden,
ändern, und zurückschreiben. Das ist bei kleinen Datenmengen völlig in Ordnung.

Interne Hilfsmethoden zum Laden und Speichern der rohen Liste:

```python
def _load_all(self) -> list[dict]:
    if not self._path.exists():
        return []
    with open(self._path, encoding="utf-8") as f:
        return json.load(f)

def _save_all(self, data: list[dict]) -> None:
    with open(self._path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)
```

`ensure_ascii=False` sorgt dafür, dass Umlaute nicht escaped werden.

Für `save_recipe`: Laden, prüfen ob Titel schon existiert, anhängen,
speichern.

Für `delete_recipe`: Laden, mit List Comprehension filtern (alle behalten, die
**nicht** den gesuchten Titel haben), prüfen ob sich die Länge geändert hat,
speichern.
