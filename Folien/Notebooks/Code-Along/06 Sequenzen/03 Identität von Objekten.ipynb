 ## Identität von Objekten

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = b
```

```python
print(f"a = {a}, b = {b}, c = {c}")
```

```python
a[0] = 10
```

```python
print(f"a = {a}, b = {b}, c = {c}")
```

```python
b[0] = 20
```

```python
c[1] = 30
```

```python
print(f"a = {a}, b = {b}, c = {c}")
```

 ## Test der Identität von Objekten

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = b
```


 `==` testet Gleichheit der Werte, nicht (notwendigerweise) Objektidentität.

```python
a == b
```

```python
b == c
```

```python
a[0] = 10
a == b
```


 Mit `is` kann man Objektidentität testen:

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = b
```

```python
a is b
```

```python
b is c
```

```python
b[0] = 10
```

```python
print(f"a = {a}, b = {b}, c = {c}")
```


 Die Funktion `id()` gibt die Adresse eines Objekts zurück:

```python
id([1, 2, 3])
```


 Meistens stellt man Adressen in hexadezimaler Form dar:

```python
hex(id([1, 2, 3]))
```
